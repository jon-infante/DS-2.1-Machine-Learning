```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

breast_cancer = load_breast_cancer()
bc = pd.DataFrame(breast_cancer.data)

bc.columns = breast_cancer.feature_names
bc['is_benign'] = breast_cancer.target
```

```python
bc.head()
```

```python
# 1 being benign
# 0 being malignant
bc['is_benign'].value_counts()
```

```python
#Getting distribution of malignant vs benign results
labels = 'Benign', 'Malignant'
fig1, ax1 = plt.subplots(figsize=(5, 5))
fig1.subplots_adjust(0.3, 0, 1, 1)
 
theme = plt.get_cmap('plasma')
ax1.set_prop_cycle("color", [theme(1. * i / len(bc['is_benign'].value_counts()))
                             for i in range(len(bc['is_benign'].value_counts()))])
 
_, _ = ax1.pie(bc['is_benign'].value_counts(), startangle=90, radius=2, labels=labels)
```
